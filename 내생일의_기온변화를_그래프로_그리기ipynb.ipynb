데이터준비하기

```python
#최고 기온 데이터만 리스트에 담아서 출력하기

import csv
f = open("/content/Seoul.csv",encoding='cp949')
data=csv.reader(f)
next(data)
result=[]

for row in data :
  if row[-1] != '': #최고 데이터 값이 존재하면 , 결측치 제외
   result.append(float(row[-1]))
print(result) # 최고기온 리스트 출력
```

데이터 개수 확인

```python
print(len(result))# 결측치 제외 된 데이터                                   
```

```python
# 최고기온 값이 8월에 해당되는 값만 추출해서 그래프 시각화
import csv
import matplotlib.pyplot as plt
f = open("/content/Seoul.csv",encoding='cp949')
data=csv.reader(f)
next(data)
result=[] # 최고기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] !='':
    if row[0].split('-')[1] == '08': # 날짜를 잘라서 08월 데이터 추출
      result.append(float(row[-1]))  # 결과 리스트에 담기

plt.figure(dpi =200) #dpi 해상도 figsize 크기
plt.plot (result,'hotpink')
plt.show()
```

매년 생일 기준 최고 기온 그래프 그리기

```python
# 최고기온 값이 8월에 해당되는 값만 추출해서 그래프 시각화
import csv
import matplotlib.pyplot as plt
f = open("/content/Seoul.csv",encoding='cp949')
data=csv.reader(f)
next(data)
result=[] # 최고기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] !='':
    if row[0].split('-')[1] == '12' and row[0].split('-')[2]=='24': # 날짜를 잘라서 08월 데이터 추출
      result.append(float(row[-1]))  # 결과 리스트에 담기

plt.figure(dpi =200) #dpi 해상도 figsize 크기
plt.plot (result,'hotpink')
plt.show()
```

매년 생일 기준 최고 기온 , 최저 기온  그래프 그리기

```python
# 최고기온 값이 8월에 해당되는 값만 추출해서 그래프 시각화
import csv
import matplotlib.pyplot as plt
f = open("/content/Seoul.csv",encoding='cp949')
data=csv.reader(f)
next(data)

high = []
low = []
for row in data :
   if row[-1] != '' and row[-2] != '':
     if row[0].split('-')[1] == '07' and row[0].split('-')[2] =='01':
       high.append(float(row[-1]))
       low.append(float(row[-2]))
plt.figure(dpi =200) #dpi 해상도 figsize 크기
plt.plot(high,'r',label='High')
plt.plot(low,'b',label='Low')
plt.legend()
plt.show()
```

날짜의 구간을 좁혀서 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt
f = open("/content/Seoul.csv",encoding='cp949')
data=csv.reader(f)
next(data)

high = []
low = []
for row in data :
   if row[-1] != '' and row[-2] != '':
     if int(row[0].split('-')[0]) >= 1997 :
       if row[0].split('-')[1] == '06' and row[0].split('-')[2] =='09':
         high.append(float(row[-1]))
         low.append(float(row[-2]))
      
plt.figure(dpi =200) #dpi 해상도 figsize 크기
plt.plot(high,'r',label='High')
plt.plot(low,'b',label='Low')
plt.legend()
plt.show()      
```

```python
import csv
import matplotlib.pyplot as plt
f = open("/content/Seoul.csv",encoding='cp949')
data=csv.reader(f)
next(data)

high = []
low = []
avg=[]
for row in data :
   if row[-1] != '' and row[-2] != '':
     if int(row[0].split('-')[0]) >= 1997 :
       if row[0].split('-')[1] == '06' and row[0].split('-')[2] =='09':
         high.append(float(row[-1]))
         low.append(float(row[-2]))
         avg.append(float(row[-3]))
#데이터 시각화
#차트 설정
plt.figure(dpi =200) #1. dpi 해상도 figsize 크기
plt.rc('font', family = 'NanumBarunGothic') #2. 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False #3. 마이너스 기호 깨짐 방지
#차트 그리기
plt.plot(high,'r',label='최고기온')
plt.plot(low,'b',label='최저기온')
plt.plot(avg,'g',label ='평균기온')
# 차트에 요소 추가
plt.legend()
plt.title("내생일 그래프")
plt.show()
```
